## Marketing Dataset
### Domain - Tourism

| **Field / Column Name**      | **Description**                                                                                      |
| ---------------------------- | ---------------------------------------------------------------------------------------------------- |
| **CustomerID**               | Unique identification number assigned to each customer                                               |
| **ProdTaken**                | Whether the customer purchased the product or not *(1 = Yes, 0 = No)*                                |
| **Age**                      | Age of the customer (in years)                                                                       |
| **TypeofContact**            | Mode of contact with the customer *(Company Invited / Self Inquiry)*                                 |
| **CityTier**                 | City classification based on development, population & living standards *(Tier 1 > Tier 2 > Tier 3)* |
| **DurationOfPitch**          | Duration (in minutes) of the sales pitch given by the salesperson                                    |
| **Occupation**               | Occupation / job profession of the customer                                                          |
| **Gender**                   | Gender of the customer                                                                               |
| **NumberOfPersonVisiting**   | Total number of persons planning to travel with the customer                                         |
| **NumberOfFollowups**        | Count of follow-ups done by salesperson after the initial pitch                                      |
| **ProductPitched**           | Product / package pitched by the salesperson                                                         |
| **PreferredPropertyStar**    | Customerâ€™s preferred hotel property rating *(e.g., 3-star / 4-star / 5-star)*                        |
| **MaritalStatus**            | Marital status of the customer *(Married / Single / Other)*                                          |
| **NumberOfTrips**            | Average number of trips the customer takes per year                                                  |
| **Passport**                 | Whether the customer has a passport *(1 = Yes, 0 = No)*                                              |
| **PitchSatisfactionScore**   | Customer satisfaction rating for the sales pitch *(Scale 1 to 5)*                                    |
| **OwnCar**                   | Whether the customer owns a car *(1 = Yes, 0 = No)*                                                  |
| **NumberOfChildrenVisiting** | Number of children below age 5 accompanying the customer                                             |
| **Designation**              | Corporate designation / job level of the customer *(e.g., AVP, Manager, Executive, etc.)*            |
| **MonthlyIncome**            | Monthly gross income of the customer (in local currency)                                             |
